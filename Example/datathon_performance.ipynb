## 1.3 Investment-Feature Multikollinearittsprfung

### Investment Feature Multicollinearity Analysis

**Business Context**: Before making investment decisions, we need to ensure that our four key features provide independent information and don't redundantly measure the same underlying customer behavior.

**Analysis Goal**: Assess whether the investment features are sufficiently independent to warrant separate investment allocations or if some features are so highly correlated that they should be considered as a single investment category.

```python
# Investment Feature Multicollinearity Analysis (VIF)
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# Focus only on the four investment features
investment_features = ['OpenFridge', 'UsedOven', 'CupsCoffee', 'AppFaceTime']

# Prepare data for VIF analysis
X_investment = data_enc[investment_features].copy()

# Check for missing values
print("=== INVESTMENT FEATURE DATA QUALITY ===")
print("Missing values per feature:")
for feature in investment_features:
    missing = X_investment[feature].isnull().sum()
    print(f"{feature}: {missing} missing values")

# Remove any rows with missing values
X_investment_clean = X_investment.dropna()
print(f"\nDataset shape after removing missing values: {X_investment_clean.shape}")

# Add constant term for VIF calculation
X_with_const = add_constant(X_investment_clean)

# Calculate VIF for each investment feature
vif_results = []
for i, feature in enumerate(investment_features):
    vif_value = variance_inflation_factor(X_with_const.values, i+1)  # +1 to skip constant
    vif_results.append({
        'Investment_Feature': feature,
        'VIF': vif_value,
        'Independence_Level': 'High' if vif_value < 2.5 else 'Moderate' if vif_value < 5 else 'Low'
    })

# Create VIF DataFrame
vif_df = pd.DataFrame(vif_results)
vif_df = vif_df.sort_values('VIF', ascending=False)

print("\n=== INVESTMENT FEATURE INDEPENDENCE ANALYSIS ===")
print("VIF (Variance Inflation Factor) Results:")
print(vif_df.to_string(index=False))

# Business interpretation
print(f"\n=== BUSINESS INTERPRETATION ===")
print("Investment Feature Independence Assessment:")

high_independence = vif_df[vif_df['Independence_Level'] == 'High']
moderate_independence = vif_df[vif_df['Independence_Level'] == 'Moderate']
low_independence = vif_df[vif_df['Independence_Level'] == 'Low']

if len(high_independence) > 0:
    print(f"\n HIGH INDEPENDENCE FEATURES ({len(high_independence)}):")
    for _, row in high_independence.iterrows():
        print(f"  - {row['Investment_Feature']}: VIF = {row['VIF']:.2f}")
    print("   These features provide unique information and warrant separate investment consideration")

if len(moderate_independence) > 0:
    print(f"\n MODERATE INDEPENDENCE FEATURES ({len(moderate_independence)}):")
    for _, row in moderate_independence.iterrows():
        print(f"  - {row['Investment_Feature']}: VIF = {row['VIF']:.2f}")
    print("   These features have some overlap but can still be considered separately")

if len(low_independence) > 0:
    print(f"\n LOW INDEPENDENCE FEATURES ({len(low_independence)}):")
    for _, row in low_independence.iterrows():
        print(f"  - {row['Investment_Feature']}: VIF = {row['VIF']:.2f}")
    print("   Consider these features as a combined investment category due to high correlation")

# Visualization
plt.figure(figsize=(12, 6))

# VIF Bar Chart
plt.subplot(1, 2, 1)
colors = ['red' if vif >= 5 else 'orange' if vif >= 2.5 else 'green' for vif in vif_df['VIF']]
bars = plt.bar(range(len(vif_df)), vif_df['VIF'], color=colors)
plt.title('Investment Feature Independence\n(VIF Analysis)', fontweight='bold')
plt.xlabel('Investment Features')
plt.ylabel('VIF Score')
plt.xticks(range(len(vif_df)), vif_df['Investment_Feature'], rotation=45)

# Add VIF interpretation lines
plt.axhline(y=2.5, color='orange', linestyle='--', alpha=0.7, label='Moderate threshold')
plt.axhline(y=5.0, color='red', linestyle='--', alpha=0.7, label='High correlation threshold')
plt.legend()

# Add values on bars
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.1, f'{height:.2f}', 
             ha='center', va='bottom', fontweight='bold')

# Correlation heatmap between investment features
plt.subplot(1, 2, 2)
feature_corr = X_investment_clean.corr()
sns.heatmap(feature_corr, annot=True, cmap='RdYlBu_r', center=0, 
            cbar_kws={'label': 'Correlation'}, fmt='.3f')
plt.title('Investment Feature Correlations', fontweight='bold')

plt.tight_layout()
plt.show()

# Summary statistics
print(f"\n=== MULTICOLLINEARITY SUMMARY ===")
avg_vif = vif_df['VIF'].mean()
max_vif = vif_df['VIF'].max()
min_vif = vif_df['VIF'].min()

print(f"Average VIF across investment features: {avg_vif:.2f}")
print(f"Highest VIF: {max_vif:.2f} ({vif_df.loc[vif_df['VIF'].idxmax(), 'Investment_Feature']})")
print(f"Lowest VIF: {min_vif:.2f} ({vif_df.loc[vif_df['VIF'].idxmin(), 'Investment_Feature']})")

if avg_vif < 2.5:
    print(" Overall assessment: Features are sufficiently independent for separate investment analysis")
elif avg_vif < 5:
    print(" Overall assessment: Moderate feature overlap - proceed with caution in investment allocation")
else:
    print(" Overall assessment: High feature overlap - consider combined investment strategy")

# Investment recommendations
print(f"\n=== INVESTMENT STRATEGY RECOMMENDATIONS ===")
print("Based on multicollinearity analysis:")

independent_features = vif_df[vif_df['VIF'] < 2.5]['Investment_Feature'].tolist()
if independent_features:
    print(f"1. INDEPENDENT INVESTMENT: {', '.join(independent_features)}")
    print(f"    Can be invested in independently without significant overlap")

correlated_features = vif_df[vif_df['VIF'] >= 5]['Investment_Feature'].tolist()
if correlated_features:
    print(f"2. COMBINED INVESTMENT: {', '.join(correlated_features)}")
    print(f"    Consider as bundled investment due to high correlation")
```

## Statistische Modellierung - Analyse der Auswirkungen von Investitionsmerkmalen

### OLS-Regression zur Quantifizierung der Investitions-ROI

**Geschftsziel**: Quantifizierung der statistischen Auswirkungen jedes Investitionsmerkmals auf die Geschftsergebnisse, um datengesttzte Empfehlungen fr die Investitionsallokation zu geben.

**Modelle**: 
1. **Umsatzmodell**: `InAppPurchase ~ OpenFridge + UsedOven + CupsCoffee + AppFaceTime`
2. **Zufriedenheitsmodell**: `UserSat ~ OpenFridge + UsedOven + CupsCoffee + AppFaceTime`  
3. **Wachstumsmodell**: `NPS ~ OpenFridge + UsedOven + CupsCoffee + AppFaceTime`

**Wichtige Kennzahlen fr Investitionsentscheidungen**:
- **Koeffizientenbetrag**: Auswirkungsgre pro Einheit Investition
- **Statistische Signifikanz**: Vertrauen in die Beziehung
- **Bestimmtheitsma (R-Quadrat)**: Erklrungsgehalt der Investitionsmerkmale
- **Geschftliche Interpretation**: Praktische Investitionsimplikationen

```python
# Investment Feature OLS Regression Analysis
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan
from scipy import stats

import pandas as pd
import numpy as np

# Load your encoded data (update the filename as needed)
# Example: data_enc = pd.read_csv('data_encoded.csv')
# If data_enc is already defined elsewhere, remove this line.
# data_enc = pd.read_csv('data_encoded.csv')

# Definiere die spezifischen Investment Features und Business Targets
investment_features = ['OpenFridge', 'UsedOven', 'CupsCoffee', 'AppFaceTime']
business_targets = ['InAppPurchase', 'UserSat', 'NPS']

# Prepare data
X = data_enc[investment_features].copy()
X = sm.add_constant(X)  # Add intercept

# Storage for results
model_results = {}
investment_rankings = {}

print("=== INVESTMENT FEATURE IMPACT ANALYSIS ===")
print("OLS Regression Results for Business Outcomes\n")

# Run regression for each business target
for target in business_targets:
    print(f"{'='*60}")
    print(f"MODEL: {target.upper()}")
    print(f"{'='*60}")
    
    # Prepare target variable
    y = data_enc[target].copy()
    
    # Fit OLS model
    model = sm.OLS(y, X).fit()
    model_results[target] = model
    
    # Display model summary
    print(f"Model Summary for {target}:")
    print(f"R-squared: {model.rsquared:.4f} ({model.rsquared*100:.1f}% variance explained)")
    print(f"Adjusted R-squared: {model.rsquared_adj:.4f}")
    print(f"F-statistic: {model.fvalue:.2f} (p-value: {model.f_pvalue:.2e})")
    print(f"Number of observations: {model.nobs}")
    
    # Extract and rank coefficients (excluding intercept)
    coefficients = model.params[1:]  # Exclude constant
    p_values = model.pvalues[1:]
    std_errors = model.bse[1:]
    
    # Create coefficient summary
    coeff_summary = pd.DataFrame({
        'Investment_Feature': investment_features,
        'Coefficient': coefficients.values,
        'Std_Error': std_errors.values,
        'P_Value': p_values.values,
        'Significance': ['***' if p < 0.001 else '**' if p < 0.01 else '*' if p < 0.05 else '' for p in p_values.values],
        'Impact_Magnitude': np.abs(coefficients.values)
    })
    
    # Sort by absolute coefficient value (impact magnitude)
    coeff_summary = coeff_summary.sort_values('Impact_Magnitude', ascending=False)
    investment_rankings[target] = coeff_summary
    
    print(f"\nInvestment Feature Impact Ranking for {target}:")
    print(coeff_summary.to_string(index=False, float_format='%.4f'))
    
    # Business interpretation
    print(f"\n BUSINESS INTERPRETATION FOR {target.upper()}:")
    
    # Top positive impact
    positive_impacts = coeff_summary[coeff_summary['Coefficient'] > 0].sort_values('Impact_Magnitude', ascending=False)
    if len(positive_impacts) > 0:
        top_positive = positive_impacts.iloc[0]
        significance = "statistically significant" if top_positive['P_Value'] < 0.05 else "not statistically significant"
        print(f" STRONGEST POSITIVE IMPACT: {top_positive['Investment_Feature']}")
        print(f"   - Coefficient: {top_positive['Coefficient']:.4f} ({significance})")
        print(f"   - Business Impact: 1 unit increase  {abs(top_positive['Coefficient']):.4f} increase in {target}")
    
    # Top negative impact (if any)
    negative_impacts = coeff_summary[coeff_summary['Coefficient'] < 0].sort_values('Impact_Magnitude', ascending=False)
    if len(negative_impacts) > 0:
        top_negative = negative_impacts.iloc[0]
        significance = "statistically significant" if top_negative['P_Value'] < 0.05 else "not statistically significant"
        print(f" STRONGEST NEGATIVE IMPACT: {top_negative['Investment_Feature']}")
        print(f"   - Coefficient: {top_negative['Coefficient']:.4f} ({significance})")
        print(f"   - Business Impact: 1 unit increase  {abs(top_negative['Coefficient']):.4f} decrease in {target}")
    
    # Significant features
    significant_features = coeff_summary[coeff_summary['P_Value'] < 0.05]
    print(f"\n STATISTICALLY SIGNIFICANT FEATURES ({len(significant_features)}):")
    for _, feature in significant_features.iterrows():
        direction = "increases" if feature['Coefficient'] > 0 else "decreases"
        print(f"   - {feature['Investment_Feature']}: {direction} {target} (p={feature['P_Value']:.4f})")
    
    print(f"\n")

# Create comprehensive investment ranking across all targets
print(f"{'='*80}")
print("COMPREHENSIVE INVESTMENT FEATURE RANKING")
print(f"{'='*80}")

# Calculate average impact across all targets
all_features_impact = []
for feature in investment_features:
    impacts = []
    significances = []
    
    for target in business_targets:
        feature_row = investment_rankings[target][investment_rankings[target]['Investment_Feature'] == feature]
        if len(feature_row) > 0:
            impacts.append(feature_row['Impact_Magnitude'].iloc[0])
            significances.append(1 if feature_row['P_Value'].iloc[0] < 0.05 else 0)
    
    avg_impact = np.mean(impacts)
    significance_count = sum(significances)
    
    all_features_impact.append({
        'Investment_Feature': feature,
        'Average_Impact_Magnitude': avg_impact,
        'Significant_Relationships': f"{significance_count}/{len(business_targets)}",
        'Investment_Priority_Score': avg_impact * (1 + significance_count * 0.2)  # Bonus for significance
    })

# Create final ranking DataFrame
final_ranking = pd.DataFrame(all_features_impact)
final_ranking = final_ranking.sort_values('Investment_Priority_Score', ascending=False)

print("Final Investment Priority Ranking:")
print(final_ranking.to_string(index=False, float_format='%.4f'))

print(f"\n STRATEGIC INVESTMENT RECOMMENDATIONS:")
print(f"Based on comprehensive statistical analysis across all business outcomes:")

for i, (_, feature) in enumerate(final_ranking.iterrows(), 1):
    priority_level = "HIGH" if i <= 2 else "MEDIUM" if i <= 3 else "LOW"
    print(f"{i}. {feature['Investment_Feature']} - {priority_level} PRIORITY")
    print(f"   Priority Score: {feature['Investment_Priority_Score']:.4f}")
    print(f"   Significant Relationships: {feature['Significant_Relationships']}")
    
    if i == 1:
        print(f"    RECOMMENDATION: Primary investment focus - highest overall impact")
    elif i == len(final_ranking):
        print(f"    RECOMMENDATION: Consider for cost optimization or feature enhancement")
    else:
        print(f"    RECOMMENDATION: Secondary investment consideration")
    print()

# Model diagnostics summary
print(f"{'='*60}")
print("MODEL PERFORMANCE SUMMARY")
print(f"{'='*60}")

for target in business_targets:
    model = model_results[target]
    print(f"{target}: R = {model.rsquared:.3f}, Adj R = {model.rsquared_adj:.3f}, F-stat = {model.fvalue:.2f}")

best_model = max(model_results.items(), key=lambda x: x[1].rsquared)
print(f"\nBest predictive model: {best_model[0]} (R = {best_model[1].rsquared:.3f})")
print(f"Investment features explain {best_model[1].rsquared*100:.1f}% of {best_model[0]} variance")
```

```python
import statsmodels.api as sm

def perform_investment_ols(data, target, features):
    X = data[features]
    X = sm.add_constant(X)
    y = data[target]
    model = sm.OLS(y, X).fit()
    return model

# Define investment features
investment_features = ['OpenFridge', 'UsedOven', 'CupsCoffee', 'AppFaceTime']

# OLS-Modelle fr alle Zielvariablen berechnen und speichern
ols_results = {}
for target in target_vars:
    ols_results[target] = perform_investment_ols(data_enc, target, investment_features)

# Investment-Modell Residuals-Analyse und Log-Transformation Prfung
fig, axes = plt.subplots(3, 4, figsize=(20, 15))

for i, target in enumerate(target_vars):
    model = ols_results[target]
    fitted_values = model.fittedvalues
    residuals = model.resid
    
    # Residuals vs Fitted
    axes[i, 0].scatter(fitted_values, residuals, alpha=0.6)
    axes[i, 0].axhline(y=0, color='red', linestyle='--')
    axes[i, 0].set_title(f'{target}: Residuals vs Fitted')
    axes[i, 0].set_xlabel('Fitted Values')
    axes[i, 0].set_ylabel('Residuals')
    
    # Q-Q Plot fr Normalitt
    stats.probplot(residuals, dist="norm", plot=axes[i, 1])
    axes[i, 1].set_title(f'{target}: Q-Q Plot')
    
    # Histogram der Residuals
    axes[i, 2].hist(residuals, bins=30, edgecolor='black', alpha=0.7)
    axes[i, 2].set_title(f'{target}: Residuals Histogram')
    axes[i, 2].set_xlabel('Residuals')
    axes[i, 2].set_ylabel('Frequency')
    
    # Log-Transformation Check (falls ntig)
    if target != 'InAppPurchase':  # Nur fr kontinuierliche Variablen
        y_values = data_enc[target]
        if y_values.min() > 0:  # Log nur mglich wenn alle Werte > 0
            y_log = np.log(y_values)
            axes[i, 3].hist(y_log, bins=30, edgecolor='black', alpha=0.7, color='orange')
            axes[i, 3].set_title(f'{target}: Log-Transformed')
            axes[i, 3].set_xlabel(f'log({target})')
            axes[i, 3].set_ylabel('Frequency')
        else:
            axes[i, 3].text(0.5, 0.5, 'Log-Transform\nnicht mglich\n(Werte  0)', 
                           ha='center', va='center', transform=axes[i, 3].transAxes)
            axes[i, 3].set_title(f'{target}: Log nicht mglich')
    else:
        axes[i, 3].text(0.5, 0.5, 'Binre Variable\nLog-Transform\nnicht relevant', 
                       ha='center', va='center', transform=axes[i, 3].transAxes)
        axes[i, 3].set_title(f'{target}: Binre Variable')

plt.tight_layout()
plt.show()

# Investment-Device Koeffizienten Extraktion und Ranking
print("\n=== INVESTMENT-DEVICE IMPACT RANKING ===")

for target in target_vars:
    model = ols_results[target]
    
    print(f"\n{target.upper()} - Investment-Device Rankings:")
    device_effects = []
    
    for feature in investment_features:
        if feature in model.params.index:
            coef = model.params[feature]
            p_val = model.pvalues[feature]
            significant = p_val < 0.05
            
            device = feature.replace('Open', '').replace('Used', '').replace('Cups', '').replace('App', 'App')
            
            device_effects.append({
                'Device': device,
                'Coefficient': coef,
                'P_Value': p_val,
                'Significant': '' if significant else '',
                'Investment_Impact': abs(coef) if significant else 0
            })
    
    # Sortiere nach Investment Impact
    device_effects.sort(key=lambda x: x['Investment_Impact'], reverse=True)
    
    for idx, effect in enumerate(device_effects, 1):
        impact_level = "HOCH" if effect['Investment_Impact'] > 0.1 else "MITTEL" if effect['Investment_Impact'] > 0.05 else "NIEDRIG"
        print(f"  {idx}. {effect['Device']:12}: Impact={effect['Coefficient']:+.4f} [{impact_level:6}] Sig={effect['Significant']} (p={effect['P_Value']:.4f})")

# Log-Transformation Empfehlung
print(f"\n{'='*60}")
print("LOG-TRANSFORMATION EMPFEHLUNG FR INVESTMENT-MODELLE")
print(f"{'='*60}")

for target in target_vars:
    if target != 'InAppPurchase':
        residuals = ols_results[target].resid
        _, shapiro_p = stats.shapiro(residuals)
        
        if shapiro_p < 0.05:
            print(f"{target}:   Log-Transformation empfohlen (p={shapiro_p:.4f})")
            
            # Teste Log-Transformation falls mglich
            y_values = data_enc[target]
            if y_values.min() > 0:
                y_log = np.log(y_values)
                model_log = perform_investment_ols(data_enc.assign(**{f'log_{target}': y_log}), f'log_{target}', investment_features)
                log_shapiro_p = stats.shapiro(model_log.resid)[1]
                
                print(f"   Log-Modell Normalitt: p={log_shapiro_p:.4f}")
                print(f"   R Original: {ols_results[target].rsquared:.4f} vs Log: {model_log.rsquared:.4f}")
                
                if log_shapiro_p > shapiro_p and model_log.rsquared > ols_results[target].rsquared:
                    print(f"    Log-Transformation verbessert Investment-Modell")
                else:
                    print(f"     Log-Transformation bringt keine Verbesserung")
        else:
            print(f"{target}:  Keine Log-Transformation ntig (p={shapiro_p:.4f})")
    else:
        print(f"{target}:   Binre Variable - Logistische Regression bereits optimal")
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.neural_network import MLPRegressor, MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, accuracy_score, classification_report
from scipy.stats import jarque_bera, shapiro
import warnings
warnings.filterwarnings('ignore')

# Daten vorbereiten
X = data_enc[investment_features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=investment_features)

# Funktion fr Regression Performance
def evaluate_regression(model, X_train, X_test, y_train, y_test, model_name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    r2 = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae = mean_absolute_error(y_test, y_pred)
    
    return {
        'Model': model_name,
        'R': r2,
        'RMSE': rmse,
        'MAE': mae,
        'fitted_model': model
    }

# Funktion fr Klassifikation Performance
def evaluate_classification(model, X_train, X_test, y_train, y_test, model_name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    
    return {
        'Model': model_name,
        'Accuracy': accuracy,
        'fitted_model': model
    }

# Initialize storage for results
results = {}
feature_importance = {}
best_models = {}

print("=== MACHINE LEARNING MODEL EVALUATION ===")

# Process each target variable
for target in target_vars:
    print(f"\n{'='*60}")
    print(f"TARGET VARIABLE: {target}")
    print(f"{'='*60}")
    
    y = data_enc[target]
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
    
    if target == 'InAppPurchase':
        # Classification models for binary target
        models = {
            'Logistic Regression': LogisticRegression(random_state=42),
            'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
            'Neural Network': MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)
        }
        
        # Evaluate classification models
        target_results = []
        for name, model in models.items():
            result = evaluate_classification(model, X_train, X_test, y_train, y_test, name)
            target_results.append(result)
            print(f"{name}:")
            print(f"  Accuracy: {result['Accuracy']:.4f}")
            
            # Store feature importance for Random Forest
            if name == 'Random Forest':
                importances = pd.DataFrame({
                    'feature': investment_features,
                    'importance': model.feature_importances_
                }).sort_values('importance', ascending=False)
                feature_importance[target] = importances
                
        results[target] = target_results
        
        # Find best model
        best_model = max(target_results, key=lambda x: x['Accuracy'])
        best_models[target] = best_model
        print(f"\nBest Model for {target}: {best_model['Model']} (Accuracy: {best_model['Accuracy']:.4f})")
        
    else:
        # Regression models for continuous targets (UserSat and NPS)
        models = {
            'Linear Regression': LinearRegression(),
            'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
            'Neural Network': MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)
        }
        
        # Evaluate regression models
        target_results = []
        for name, model in models.items():
            result = evaluate_regression(model, X_train, X_test, y_train, y_test, name)
            target_results.append(result)
            print(f"{name}:")
            print(f"  R: {result['R']:.4f}")
            print(f"  RMSE: {result['RMSE']:.4f}")
            print(f"  MAE: {result['MAE']:.4f}")
            
            # Store feature importance for Random Forest
            if name == 'Random Forest':
                importances = pd.DataFrame({
                    'feature': investment_features,
                    'importance': model.feature_importances_
                }).sort_values('importance', ascending=False)
                feature_importance[target] = importances
        
        results[target] = target_results
        
        # Find best model
        best_model = max(target_results, key=lambda x: x['R'])
        best_models[target] = best_model
        print(f"\nBest Model for {target}: {best_model['Model']} (R: {best_model['R']:.4f})")

# Combined Feature Importance Analysis
print(f"\n{'='*60}")
print("COMPREHENSIVE FEATURE IMPORTANCE ANALYSIS")
print(f"{'='*60}")

# Combine importance scores across all targets
combined_importance = pd.DataFrame()
for target, importance_df in feature_importance.items():
    combined_importance[target] = importance_df.set_index('feature')['importance']

# Calculate overall importance
combined_importance['Average_Importance'] = combined_importance.mean(axis=1)
combined_importance = combined_importance.sort_values('Average_Importance', ascending=False)

print("\nOverall Feature Importance Ranking:")
print(combined_importance)

# Visualization of feature importance
plt.figure(figsize=(15, 10))

# Subplot 1: Overall Feature Importance
plt.subplot(2, 1, 1)
combined_importance['Average_Importance'].plot(kind='bar')
plt.title('Overall Feature Importance Across All Targets')
plt.xlabel('Features')
plt.ylabel('Average Importance Score')
plt.xticks(rotation=45)

# Subplot 2: Feature Importance by Target
plt.subplot(2, 1, 2)
combined_importance.drop('Average_Importance', axis=1).plot(kind='bar')
plt.title('Feature Importance by Target Variable')
plt.xlabel('Features')
plt.ylabel('Importance Score')
plt.legend(title='Target Variables')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# Investment recommendations based on ML insights
print(f"\n{'='*60}")
print("INVESTMENT RECOMMENDATIONS FROM ML ANALYSIS")
print(f"{'='*60}")

# Get top features for each target
for target in target_vars:
    importance = feature_importance[target]
    top_feature = importance.iloc[0]
    print(f"\n{target} - Top Investment Priority:")
    print(f"  Feature: {top_feature['feature']}")
    print(f"  Importance Score: {top_feature['importance']:.4f}")
    
    # Model performance summary
    best_model = best_models[target]
    if target == 'InAppPurchase':
        print(f"  Model Performance: {best_model['Model']} (Accuracy: {best_model['Accuracy']:.4f})")
    else:
        print(f"  Model Performance: {best_model['Model']} (R: {best_model['R']:.4f})")

# Overall investment strategy
print(f"\n{'='*60}")
print("OVERALL INVESTMENT STRATEGY")
print(f"{'='*60}")

overall_importance = combined_importance['Average_Importance']
for i, (feature, importance) in enumerate(overall_importance.items(), 1):
    if i == 1:
        priority = "HIGH"
        budget = "Primary Investment"
    elif i == 2:
        priority = "HIGH-MEDIUM"
        budget = "Secondary Investment"
    elif i == 3:
        priority = "MEDIUM"
        budget = "Moderate Investment"
    else:
        priority = "LOW"
        budget = "Minimal Investment"
        
    print(f"\n{i}. {feature}:")
    print(f"   Priority: {priority}")
    print(f"   Budget Allocation: {budget}")
    print(f"   Overall Importance Score: {importance:.4f}")
    
    # Add target-specific insights
    print(f"   Target Impact:")
    for target in target_vars:
        target_importance = feature_importance[target]
        target_score = target_importance[target_importance['feature'] == feature]['importance'].iloc[0]
        print(f"     - {target}: {target_score:.4f}")
```

```python
# Advanced Machine Learning for Investment Feature Analysis
from sklearn.model_selection import train_test_split, cross_val_score, learning_curve
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import xgboost as xgb
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.neural_network import MLPRegressor, MLPClassifier

# Definiere die vier spezifischen Investment Features
investment_features = ['OpenFridge', 'UsedOven', 'CupsCoffee', 'AppFaceTime']
business_targets = ['InAppPurchase', 'UserSat', 'NPS']

# Prepare data
X = data_enc[investment_features].copy()
print(f"Investment feature matrix shape: {X.shape}")
print(f"Investment features: {list(X.columns)}")

# Scale features for ML models
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=investment_features)

# Initialize models
models = {
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'XGBoost': xgb.XGBRegressor(random_state=42),
    'Neural Network': MLPRegressor(hidden_layer_sizes=(20, 10), max_iter=1000, random_state=42)
}

# Store results
results = {}
feature_importance = {}

for target in business_targets:
    print(f"\n{'='*60}")
    print(f"TARGET: {target.upper()}")
    print(f"{'='*60}")
    
    y = data_enc[target].copy()
    
    if target == 'InAppPurchase':
        # Klassifikation fr binre Variable
        models = {
            'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
            'XGBoost': xgb.XGBClassifier(random_state=42),
            'Neural Network': MLPClassifier(hidden_layer_sizes=(20, 10), max_iter=1000, random_state=42)
        }
    elif target in ['UserSat', 'NPS']:
        # Shift classes to start from 0 for classifiers
        y = y - y.min()
    
    # Train-Test Split
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
    
    results[target] = {}
    feature_importance[target] = {}
    
    for name, model in models.items():
        print(f"\n Training {name}...")
        
        # Train model
        model.fit(X_train, y_train)
        
        # Predictions
        y_pred = model.predict(X_test)
        
        # Calculate metrics
        if target == 'InAppPurchase':
            from sklearn.metrics import accuracy_score, classification_report
            accuracy = accuracy_score(y_test, y_pred)
            print(f"Test Accuracy: {accuracy:.4f}")
            print("\nClassification Report:")
            print(classification_report(y_test, y_pred))
        else:
            r2 = r2_score(y_test, y_pred)
            rmse = np.sqrt(mean_squared_error(y_test, y_pred))
            print(f"Test R: {r2:.4f}")
            print(f"Test RMSE: {rmse:.4f}")
        
        # Feature importance
        if hasattr(model, 'feature_importances_'):
            importances = model.feature_importances_
        elif hasattr(model, 'coef_'):
            importances = np.abs(model.coef_)
        else:
            importances = np.zeros(len(investment_features))
            
        feature_importance[target][name] = dict(zip(investment_features, importances))
        
        # Plot feature importance
        plt.figure(figsize=(10, 4))
        features = investment_features
        if hasattr(model, 'feature_importances_'):
            importance = model.feature_importances_
        elif hasattr(model, 'coef_'):
            importance = np.abs(model.coef_)
        else:
            continue
            
        indices = np.argsort(importance)[::-1]
        plt.title(f'Feature Importance ({name} - {target})')
        plt.bar(range(len(indices)), importance[indices])
        plt.xticks(range(len(indices)), [features[i] for i in indices], rotation=45)
        plt.tight_layout()
        plt.show()

# Summarize feature importance across all models and targets
print("\n" + "="*60)
print("OVERALL FEATURE IMPORTANCE SUMMARY")
print("="*60)

overall_importance = pd.DataFrame(index=investment_features)

for target in business_targets:
    for model_name in models.keys():
        importance = feature_importance[target][model_name]
        col_name = f"{target}_{model_name}"
        overall_importance[col_name] = pd.Series(importance)

# Calculate mean importance across all models and targets
overall_importance['Mean_Importance'] = overall_importance.mean(axis=1)
overall_importance = overall_importance.sort_values('Mean_Importance', ascending=False)

print("\nOverall Feature Ranking:")
for feature in overall_importance.index:
    mean_imp = overall_importance.loc[feature, 'Mean_Importance']
    print(f"{feature}: {mean_imp:.4f}")
    
# Visualize overall importance
plt.figure(figsize=(10, 4))
plt.title('Overall Feature Importance')
plt.bar(range(len(investment_features)), overall_importance['Mean_Importance'])
plt.xticks(range(len(investment_features)), overall_importance.index, rotation=45)
plt.tight_layout()
plt.show()

# Print investment recommendations
print("\n" + "="*60)
print("INVESTMENT RECOMMENDATIONS")
print("="*60)

for i, (feature, row) in enumerate(overall_importance.iterrows(), 1):
    importance = row['Mean_Importance']
    if i == 1:
        priority = "HIGHEST"
        budget = "40-50%"
    elif i == 2:
        priority = "HIGH"
        budget = "25-30%"
    elif i == 3:
        priority = "MEDIUM"
        budget = "15-20%"
    else:
        priority = "LOW"
        budget = "5-10%"
        
    print(f"\n{i}. {feature}")
    print(f"   Priority: {priority}")
    print(f"   Recommended Budget Allocation: {budget}")
    print(f"   Relative Importance Score: {importance:.4f}")
```
