```python
# Clustering durchfhren
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(clustering_data_scaled)

# Clusterlabels zu den Daten hinzufgen
data_new['Cluster'] = clusters
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
clustering_data_pca = pca.fit_transform(clustering_data_scaled)

# Visualisierung der Cluster
plt.figure(figsize=(10, 6))
plt.scatter(clustering_data_pca[:, 0], clustering_data_pca[:, 1], c=clusters, cmap='viridis', alpha=0.6)
plt.title('Visualisierung der Cluster', fontsize=16)
plt.xlabel('PCA-Komponente 1', fontsize=14)
plt.ylabel('PCA-Komponente 2', fontsize=14)
plt.colorbar(label='Cluster')
plt.show()
```
