```python
from statsmodels.api import add_constant

target = 'SalesVol'
 
numeric_features = data_clean.select_dtypes(include=['float64', 'int64']).drop(columns=[target], errors='ignore')
# Add constant (intercept) term
df_const = add_constant(numeric_features)
 
# Compute VIF for each feature
vif_data = pd.DataFrame({
    'Feature': df_const.columns,
    'VIF': [
        variance_inflation_factor(df_const.values, i)
        for i in range(df_const.shape[1])
    ]
})
 
print(vif_data)
```

## Feature-Auswahl Empfehlung basierend auf VIF & Korrelationsanalyse

### Analyseergebnisse:

**VIF-Kategorisierung:**
-  **Kritisch (VIF > 10)**: OpenFridge (12.16), NPS (11.17)
-  **Moderat (VIF 5-10)**: UserSat (8.35), Product3 (5.77)
-  **Akzeptabel (VIF < 5)**: Alle anderen Features

**Korrelationserkenntnisse:**
- **Strkste SalesVol-Prdiktoren**: AppFaceTime (0.83), Individuals (0.78)
- **Strkste InAppPurchase-Prdiktoren**: UsedOven (0.79), Product3 (0.71)
- **Schwache Prdiktoren**: CupsCoffee, Product1, Product2, Weather Features

```python
# FEATURE SELECTION STRATEGIE basierend auf VIF & Korrelation
print("=== FEATURE SELECTION EMPFEHLUNG ===")

# Sortiere VIF Daten ohne const fr bessere bersicht
vif_analysis = vif_data[vif_data['Feature'] != 'const'].sort_values('VIF', ascending=False)
print("VIF Ranking (ohne Konstante):")
print(vif_analysis)

# Definiere Ausschlusskriterien
excluded_features = []
selected_features = []

# 1. AUSSCHLUSS: Features mit VIF > 10
high_vif = vif_analysis[vif_analysis['VIF'] > 10]['Feature'].tolist()
excluded_features.extend(high_vif)
print(f"\n AUSGESCHLOSSEN (VIF > 10): {high_vif}")

# 2. AUSSCHLUSS: Features mit sehr schwacher Korrelation (< 0.1 fr beide Ziele)
correlation_data = correlation_with_targets.drop(['SalesVol', 'InAppPurchase'])
weak_corr_features = []
for feature in correlation_data.index:
    if feature in data_clean.columns:  # Nur numerische Features betrachten
        sales_corr = abs(correlation_data.loc[feature, 'SalesVol'])
        purchase_corr = abs(correlation_data.loc[feature, 'InAppPurchase'])
        if sales_corr < 0.1 and purchase_corr < 0.1:
            weak_corr_features.append(feature)

excluded_features.extend(weak_corr_features)
print(f" AUSGESCHLOSSEN (schwache Korrelation < 0.1): {weak_corr_features}")

# 3. EINSCHLUSS: Verbliebene Features bewerten
remaining_features = vif_analysis[~vif_analysis['Feature'].isin(excluded_features)]

print(f"\n VERBLEIBENDE FEATURES FR MODELLIERUNG:")
for _, row in remaining_features.iterrows():
    feature = row['Feature']
    vif_val = row['VIF']
    
    # Korrelationswerte holen (falls verfgbar)
    if feature in correlation_data.index:
        sales_corr = correlation_data.loc[feature, 'SalesVol']
        purchase_corr = correlation_data.loc[feature, 'InAppPurchase']
        print(f"   {feature}: VIF={vif_val:.2f}, SalesVol_corr={sales_corr:.3f}, InApp_corr={purchase_corr:.3f}")
        selected_features.append(feature)
    else:
        print(f"   {feature}: VIF={vif_val:.2f} (kategorisches Feature)")
        selected_features.append(feature)

# Fge wichtige kategorische Features hinzu
categorical_features = ['Weather_Rainy', 'Weather_Sunny', 'Region_NL', 'Region_UK']
available_categorical = [f for f in categorical_features if f in data_enc.columns]
selected_features.extend(available_categorical)

print(f"\n FINALE FEATURE AUSWAHL:")
print(f"Anzahl Features: {len(selected_features)}")
print(f"Features: {selected_features}")

# Erstelle finalen Datensatz
X_selected = data_enc[selected_features]
print(f"\nDataset Shape: {X_selected.shape}")
print(f"Features im finalen Dataset: {list(X_selected.columns)}")
```

## 3. Logistische Regression mit SalesVol als Zielvariable

### Problemdefinition
Da SalesVol eine kontinuierliche Variable ist, werden wir sie fr die logistische Regression in eine binre Klassifikation umwandeln:
- **High Sales** (1): SalesVol ber dem Median
- **Low Sales** (0): SalesVol unter oder gleich dem Median

```python
# SalesVol in binre Zielvariable umwandeln
print("=== SALESVOL BINARISIERUNG ===")

# Analysiere SalesVol Verteilung
salesvol_median = data_enc['SalesVol'].median()
salesvol_mean = data_enc['SalesVol'].mean()
salesvol_q75 = data_enc['SalesVol'].quantile(0.75)

print(f"SalesVol Statistiken:")
print(f"   Median: {salesvol_median:.2f}")
print(f"   Mean: {salesvol_mean:.2f}")
print(f"   75% Quantil: {salesvol_q75:.2f}")

# Erstelle binre Zielvariable (1 = High Sales, 0 = Low Sales)
# Verwende Median als Schwellenwert fr ausgewogene Klassen
data_enc['SalesVol_Binary'] = (data_enc['SalesVol'] > salesvol_median).astype(int)

print(f"\n=== BINRE ZIELVARIABLE ===")
print(f"Schwellenwert: {salesvol_median:.2f}")
print("Klassenverteilung:")
class_distribution = data_enc['SalesVol_Binary'].value_counts(normalize=True)
print(f"   Low Sales (0): {class_distribution[0]:.1%}")
print(f"   High Sales (1): {class_distribution[1]:.1%}")

# Visualisierung der Binarisierung
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Original SalesVol Distribution
ax1.hist(data_enc['SalesVol'], bins=50, alpha=0.7, edgecolor='black')
ax1.axvline(salesvol_median, color='red', linestyle='--', linewidth=2, label=f'Median ({salesvol_median:.0f})')
ax1.set_title('SalesVol Verteilung mit Median-Schwellenwert')
ax1.set_xlabel('SalesVol')
ax1.set_ylabel('Hufigkeit')
ax1.legend()

# Binary Target Distribution
class_counts = data_enc['SalesVol_Binary'].value_counts()
ax2.bar(['Low Sales (0)', 'High Sales (1)'], class_counts.values, color=['lightcoral', 'lightgreen'])
ax2.set_title('Binre Zielvariable Verteilung')
ax2.set_ylabel('Anzahl Beobachtungen')
for i, v in enumerate(class_counts.values):
    ax2.text(i, v + 20, str(v), ha='center', va='bottom')

plt.tight_layout()
plt.show()

print(f"\n Binre Zielvariable 'SalesVol_Binary' erfolgreich erstellt")
```

```python
# Finale Feature-Auswahl fr Logistische Regression
print("=== FINALE FEATURE AUSWAHL ===")

# Entferne SalesVol (Original) und SalesVol_Binary (Ziel) aus den Features
final_features = [f for f in selected_features if f not in ['SalesVol', 'SalesVol_Binary']]
print(f"Ausgewhlte Features: {final_features}")
print(f"Anzahl Features: {len(final_features)}")

# Erstelle Feature Matrix und Zielvariable
X_features = data_enc[final_features]
y_target = data_enc['SalesVol_Binary']

print(f"\nFeature Matrix Shape: {X_features.shape}")
print(f"Target Variable Shape: {y_target.shape}")

# Zeige Feature-Korrelationen mit der ursprnglichen SalesVol
print(f"\n=== FEATURE KORRELATIONEN MIT SALESVOL ===")
feature_correlations = data_enc[final_features + ['SalesVol']].corr()['SalesVol'].drop('SalesVol').sort_values(key=abs, ascending=False)
for feature, corr in feature_correlations.items():
    print(f"   {feature}: {corr:.3f}")

# Korrelationsmatrix der finalen Features
print(f"\n=== KORRELATIONSMATRIX DER FINALEN FEATURES ===")

# Erstelle Korrelationsmatrix
correlation_matrix_final = X_features.corr()

# Visualisierung
plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(correlation_matrix_final, dtype=bool))
sns.heatmap(correlation_matrix_final, 
            mask=mask,
            annot=True, 
            fmt=".2f", 
            cmap="RdBu_r", 
            center=0,
            square=True,
            cbar_kws={"shrink": .8})
plt.title('Korrelationsmatrix der Features fr SalesVol Logistic Regression', fontsize=14, pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Prfe auf hohe Korrelationen zwischen Features
print(f"\n=== MULTIKOLLINEARITTSPRFUNG ===")
high_corr_pairs = []
for i in range(len(correlation_matrix_final.columns)):
    for j in range(i+1, len(correlation_matrix_final.columns)):
        corr_val = correlation_matrix_final.iloc[i, j]
        if abs(corr_val) > 0.7:
            high_corr_pairs.append((correlation_matrix_final.columns[i], correlation_matrix_final.columns[j], corr_val))

if high_corr_pairs:
    print("  Hohe Korrelationen zwischen Features gefunden:")
    for feat1, feat2, corr in high_corr_pairs:
        print(f"   {feat1}  {feat2}: {corr:.3f}")
else:
    print(" Keine kritischen Korrelationen zwischen Features (|r| > 0.7)")
```

```python
# Train-Test Split
print("=== TRAIN-TEST SPLIT ===")

# Aufteilen der Daten (80% Training, 20% Test)
X_train, X_test, y_train, y_test = train_test_split(
    X_features, y_target, 
    test_size=0.2, 
    random_state=42, 
    stratify=y_target  # Sorgt fr ausgewogene Klassenverteilung
)

print(f"Training Set: {X_train.shape[0]} Beobachtungen")
print(f"Test Set: {X_test.shape[0]} Beobachtungen")
print(f"Anzahl Features: {X_train.shape[1]}")

# berprfe Klassenverteilung nach Split
print(f"\n=== KLASSENVERTEILUNG NACH SPLIT ===")
print("Training Set:")
train_dist = y_train.value_counts(normalize=True)
print(f"   Low Sales (0): {train_dist[0]:.1%}")
print(f"   High Sales (1): {train_dist[1]:.1%}")

print("\nTest Set:")
test_dist = y_test.value_counts(normalize=True)
print(f"   Low Sales (0): {test_dist[0]:.1%}")
print(f"   High Sales (1): {test_dist[1]:.1%}")

# Feature Standardisierung
print(f"\n=== FEATURE STANDARDISIERUNG ===")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Konvertiere zurck zu DataFrame
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)

print(" Features standardisiert (Mean0, Std1)")
print(f"Training - Mean: {X_train_scaled.mean().mean():.4f}, Std: {X_train_scaled.std().mean():.4f}")
print(f"Test - Mean: {X_test_scaled.mean().mean():.4f}, Std: {X_test_scaled.std().mean():.4f}")

# Datenbersicht
print(f"\n=== DATENBERSICHT ===")
print("Feature Namen:")
for i, feature in enumerate(X_train.columns, 1):
    print(f"  {i:2d}. {feature}")

print(f"\nErste 3 Zeilen der standardisierten Trainingsdaten:")
print(X_train_scaled.head(3).round(3))
```

```python
# Logistische Regression Training
print("=== LOGISTISCHE REGRESSION TRAINING ===")

# Initialisiere und trainiere das Modell
log_reg = LogisticRegression(
    random_state=42,
    max_iter=1000,
    solver='liblinear',  # Gut fr kleinere Datasets
    C=1.0  # Regularisierungsparameter
)

# Trainiere das Modell
log_reg.fit(X_train_scaled, y_train)
print(" Logistische Regression erfolgreich trainiert")

# Vorhersagen
y_train_pred = log_reg.predict(X_train_scaled)
y_test_pred = log_reg.predict(X_test_scaled)
y_train_proba = log_reg.predict_proba(X_train_scaled)[:, 1]
y_test_proba = log_reg.predict_proba(X_test_scaled)[:, 1]

# Performance Metriken
print(f"\n=== MODELL PERFORMANCE ===")
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Training Metriken
train_accuracy = accuracy_score(y_train, y_train_pred)
train_precision = precision_score(y_train, y_train_pred)
train_recall = recall_score(y_train, y_train_pred)
train_f1 = f1_score(y_train, y_train_pred)
train_auc = roc_auc_score(y_train, y_train_proba)

# Test Metriken
test_accuracy = accuracy_score(y_test, y_test_pred)
test_precision = precision_score(y_test, y_test_pred)
test_recall = recall_score(y_test, y_test_pred)
test_f1 = f1_score(y_test, y_test_pred)
test_auc = roc_auc_score(y_test, y_test_proba)

# Performance Tabelle
performance_df = pd.DataFrame({
    'Metrik': ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'AUC-ROC'],
    'Training': [train_accuracy, train_precision, train_recall, train_f1, train_auc],
    'Test': [test_accuracy, test_precision, test_recall, test_f1, test_auc],
    'Differenz': [train_accuracy-test_accuracy, train_precision-test_precision, 
                  train_recall-test_recall, train_f1-test_f1, train_auc-test_auc]
})

print(performance_df.round(4))

# Overfitting/Underfitting Analyse
print(f"\n=== GENERALISIERUNG ANALYSE ===")
accuracy_diff = train_accuracy - test_accuracy
if accuracy_diff > 0.05:
    print(f"  Potentielles Overfitting erkannt (Differenz: {accuracy_diff:.4f})")
elif accuracy_diff < -0.02:
    print(f"  Potentielles Underfitting erkannt (Differenz: {accuracy_diff:.4f})")
else:
    print(f" Gute Generalisierung (Differenz: {accuracy_diff:.4f})")

# Feature Importance (Koeffizienten)
print(f"\n=== FEATURE IMPORTANCE ===")
feature_importance = pd.DataFrame({
    'Feature': X_train.columns,
    'Koeffizient': log_reg.coef_[0],
    'Abs_Koeffizient': np.abs(log_reg.coef_[0])
}).sort_values('Abs_Koeffizient', ascending=False)

print("Top 10 wichtigste Features:")
print(feature_importance.head(10).round(4))
```

## 4. Random Forest Klassifikation

### Modell-Rationale
Random Forest kann Overfitting besser handhaben durch:
- **Ensemble-Methode**: Kombiniert viele Entscheidungsbume
- **Bootstrap Sampling**: Reduziert Varianz
- **Feature Randomness**: Verringert beranpassung an spezifische Features
- **Built-in Regularisierung**: Durch max_depth, min_samples_split, etc.

```python
# Random Forest Classifier Implementation
print("=== Random Forest Classifier ===")

# Import additional libraries for Random Forest
from sklearn.model_selection import GridSearchCV
import time

# Initialize Random Forest with initial parameters
rf_classifier = RandomForestClassifier(
    random_state=42,
    n_jobs=-1  # Use all available cores
)

# Hyperparameter tuning with GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 15, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

print("Starting hyperparameter tuning for Random Forest...")
start_time = time.time()

# Perform grid search with cross-validation
rf_grid_search = GridSearchCV(
    estimator=rf_classifier,
    param_grid=param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)

# Fit the grid search
rf_grid_search.fit(X_train_scaled, y_train)

end_time = time.time()
print(f"Grid search completed in {end_time - start_time:.2f} seconds")

# Best parameters
print("\nBest Random Forest parameters:")
print(rf_grid_search.best_params_)
print(f"Best cross-validation ROC-AUC score: {rf_grid_search.best_score_:.4f}")

# Use the best estimator
rf_best = rf_grid_search.best_estimator_
```

```python
# Random Forest Model Evaluation
print("\n=== Random Forest Model Evaluation ===")

# Make predictions
y_train_pred_rf = rf_best.predict(X_train_scaled)
y_test_pred_rf = rf_best.predict(X_test_scaled)
y_train_proba_rf = rf_best.predict_proba(X_train_scaled)[:, 1]
y_test_proba_rf = rf_best.predict_proba(X_test_scaled)[:, 1]

# Calculate performance metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Training performance
train_accuracy_rf = accuracy_score(y_train, y_train_pred_rf)
train_precision_rf = precision_score(y_train, y_train_pred_rf)
train_recall_rf = recall_score(y_train, y_train_pred_rf)
train_f1_rf = f1_score(y_train, y_train_pred_rf)
train_auc_rf = roc_auc_score(y_train, y_train_proba_rf)

# Test performance
test_accuracy_rf = accuracy_score(y_test, y_test_pred_rf)
test_precision_rf = precision_score(y_test, y_test_pred_rf)
test_recall_rf = recall_score(y_test, y_test_pred_rf)
test_f1_rf = f1_score(y_test, y_test_pred_rf)
test_auc_rf = roc_auc_score(y_test, y_test_proba_rf)

# Performance comparison
print("Random Forest Performance:")
print(f"Training - Accuracy: {train_accuracy_rf:.4f}, Precision: {train_precision_rf:.4f}, Recall: {train_recall_rf:.4f}, F1: {train_f1_rf:.4f}, AUC: {train_auc_rf:.4f}")
print(f"Test     - Accuracy: {test_accuracy_rf:.4f}, Precision: {test_precision_rf:.4f}, Recall: {test_recall_rf:.4f}, F1: {test_f1_rf:.4f}, AUC: {test_auc_rf:.4f}")

# Calculate overfitting indicators
accuracy_diff_rf = train_accuracy_rf - test_accuracy_rf
auc_diff_rf = train_auc_rf - test_auc_rf

print(f"\nOverfitting Analysis:")
print(f"Accuracy difference (Train - Test): {accuracy_diff_rf:.4f}")
print(f"AUC difference (Train - Test): {auc_diff_rf:.4f}")

if accuracy_diff_rf < 0.05:
    print(" Good generalization - Low overfitting")
elif accuracy_diff_rf < 0.1:
    print(" Moderate overfitting")
else:
    print(" High overfitting")

# Create performance comparison DataFrame
rf_performance = pd.DataFrame({
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'ROC-AUC'],
    'Training': [train_accuracy_rf, train_precision_rf, train_recall_rf, train_f1_rf, train_auc_rf],
    'Test': [test_accuracy_rf, test_precision_rf, test_recall_rf, test_f1_rf, test_auc_rf]
})

print("\nDetailed Performance Metrics:")
print(rf_performance.round(4))
```

## 5. XGBoost Klassifikation

XGBoost (eXtreme Gradient Boosting) ist ein hochperformantes Gradient Boosting Framework, das fr seine Effizienz und Genauigkeit bekannt ist. Im Gegensatz zu Random Forest, das viele Bume parallel trainiert, baut XGBoost Bume sequenziell auf und korrigiert dabei die Fehler der vorherigen Modelle.

**Vorteile von XGBoost:**
- Hohe Vorhersagegenauigkeit durch Ensemble-Learning
- Integrierte Regularisierung zur Vermeidung von Overfitting
- Effiziente Behandlung von Missing Values
- Feature Importance basierend auf Gain/Cover/Frequency

**Erwartung:** XGBoost sollte hnlich gute Ergebnisse wie Random Forest liefern, mglicherweise mit noch besserer Generalisierung durch die eingebaute Regularisierung.

```python
# XGBoost Implementation
print("=== XGBoost Classifier ===")

# Install and import XGBoost
try:
    import xgboost as xgb
    print("XGBoost is already installed")
except ImportError:
    print("Installing XGBoost...")
    import subprocess
    import sys
    subprocess.check_call([sys.executable, "-m", "pip", "install", "xgboost"])
    import xgboost as xgb
    print("XGBoost installed successfully")

# Initialize XGBoost classifier
xgb_classifier = xgb.XGBClassifier(
    random_state=42,
    n_jobs=-1,
    eval_metric='logloss'  # Suppress warning
)

# Hyperparameter tuning for XGBoost
xgb_param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 6, 9, 12],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0]
}

print("Starting hyperparameter tuning for XGBoost...")
start_time = time.time()

# Perform grid search with cross-validation
xgb_grid_search = GridSearchCV(
    estimator=xgb_classifier,
    param_grid=xgb_param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1,
    verbose=1
)

# Fit the grid search
xgb_grid_search.fit(X_train_scaled, y_train)

end_time = time.time()
print(f"Grid search completed in {end_time - start_time:.2f} seconds")

# Best parameters
print("\nBest XGBoost parameters:")
print(xgb_grid_search.best_params_)
print(f"Best cross-validation ROC-AUC score: {xgb_grid_search.best_score_:.4f}")

# Use the best estimator
xgb_best = xgb_grid_search.best_estimator_
```

```python
# XGBoost Model Evaluation
print("\n=== XGBoost Model Evaluation ===")

# Make predictions
y_train_pred_xgb = xgb_best.predict(X_train_scaled)
y_test_pred_xgb = xgb_best.predict(X_test_scaled)
y_train_proba_xgb = xgb_best.predict_proba(X_train_scaled)[:, 1]
y_test_proba_xgb = xgb_best.predict_proba(X_test_scaled)[:, 1]

# Calculate performance metrics
# Training performance
train_accuracy_xgb = accuracy_score(y_train, y_train_pred_xgb)
train_precision_xgb = precision_score(y_train, y_train_pred_xgb)
train_recall_xgb = recall_score(y_train, y_train_pred_xgb)
train_f1_xgb = f1_score(y_train, y_train_pred_xgb)
train_auc_xgb = roc_auc_score(y_train, y_train_proba_xgb)

# Test performance
test_accuracy_xgb = accuracy_score(y_test, y_test_pred_xgb)
test_precision_xgb = precision_score(y_test, y_test_pred_xgb)
test_recall_xgb = recall_score(y_test, y_test_pred_xgb)
test_f1_xgb = f1_score(y_test, y_test_pred_xgb)
test_auc_xgb = roc_auc_score(y_test, y_test_proba_xgb)

# Performance comparison
print("XGBoost Performance:")
print(f"Training - Accuracy: {train_accuracy_xgb:.4f}, Precision: {train_precision_xgb:.4f}, Recall: {train_recall_xgb:.4f}, F1: {train_f1_xgb:.4f}, AUC: {train_auc_xgb:.4f}")
print(f"Test     - Accuracy: {test_accuracy_xgb:.4f}, Precision: {test_precision_xgb:.4f}, Recall: {test_recall_xgb:.4f}, F1: {test_f1_xgb:.4f}, AUC: {test_auc_xgb:.4f}")

# Calculate overfitting indicators
accuracy_diff_xgb = train_accuracy_xgb - test_accuracy_xgb
auc_diff_xgb = train_auc_xgb - test_auc_xgb

print(f"\nOverfitting Analysis:")
print(f"Accuracy difference (Train - Test): {accuracy_diff_xgb:.4f}")
print(f"AUC difference (Train - Test): {auc_diff_xgb:.4f}")

if accuracy_diff_xgb < 0.05:
    print(" Good generalization - Low overfitting")
elif accuracy_diff_xgb < 0.1:
    print(" Moderate overfitting")
else:
    print(" High overfitting")

# Create performance comparison DataFrame
xgb_performance = pd.DataFrame({
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'ROC-AUC'],
    'Training': [train_accuracy_xgb, train_precision_xgb, train_recall_xgb, train_f1_xgb, train_auc_xgb],
    'Test': [test_accuracy_xgb, test_precision_xgb, test_recall_xgb, test_f1_xgb, test_auc_xgb]
})

print("\nDetailed Performance Metrics:")
print(xgb_performance.round(4))
```
